```python
# Data provided by Redfin, a national real estate brokerage
import pandas as pd
```

```python
# Store filepath in a variable
raw_data = "Resources/weekly_housing_market_rawdata_10.10.2020.csv"

# Read our Data file with the pandas library
raw_data_df = pd.read_csv(raw_data, encoding="utf-8")
raw_data_df.head()
```

```python
# Show just the header
len(raw_data_df)
```

```python
raw_data_df.columns
```

```python
columns1=[
    "region_name", 
    "region_type", 
    "period_begin", 
    "period_end",
    "total_homes_sold", 
    "total_new_listings", 
    "total_active_listings",
    "median_days_on_market",
    "months_of_supply",
    "age_of_inventory"]

weekly_df=raw_data_df.loc[raw_data_df["duration"] == "1 weeks",columns1]
weekly_df.head()
```

```python
len(weekly_df)
```

```python
weekly_df.dtypes
```

```python
#start a new DataFrame to cleanup
clean_df=weekly_df

#break out period_begin into month, day, year
clean_df[['month','day','year']] = clean_df.period_begin.str.split("/",expand=True)

#add leading 20 to year
clean_df['year'] ='20' + clean_df['year']

#replace month number with month name
clean_df['month']=clean_df['month'].replace(
    {'1':'M01',
    '2':'M02',
    '3':'M03',
    '4':'M04',
    '5':'M05',
    '6':'M06',
    '7':'M07',
    '8':'M08',
    '9':'M09',
    '10':'M10',
    '11':'M11',
    '12':'M12'})

#merge month and year for grouping
clean_df['month_year'] =clean_df['month'].astype(str) + " "+ clean_df['year']
clean_df.head()
```

```python
#break out region_name into city, and state
clean_df[['region','state']] = clean_df.region_name.str.split(", ",expand=True)

#only take first two characters of state
clean_df['state'] = clean_df['state'].str[:2]
clean_df.head()
```

```python
clean_df.year.value_counts()
```

```python
# We want to separate out RedFin data that will be summed when grouped
sum_re_df = clean_df[["month", "year","month_year","region", "state",
                            "total_homes_sold", "total_new_listings", "total_active_listings"]]
sum_re_df.head()
```

```python
# Using GroupBy in order to separate the data into fields according to "month_year" values
grouped_sum_df = sum_re_df.groupby(['month_year'])
grouped_sum_df.sum().head()
```

```python
# We want to separate out RedFin data that will be averaged when grouped
avg_re_df=clean_df[["month", "year","month_year","region", "state",
                            "median_days_on_market","months_of_supply","age_of_inventory"]]
avg_re_df.head()
```

```python
# Using GroupBy in order to separate the data into fields according to "month_year" values
grouped_avg_df = avg_re_df.groupby(['month_year'])
grouped_avg_df.mean().head()
```

```python
# Merge two dataframes using an inner join
real_estate_df = pd.merge(grouped_avg_df.mean(), grouped_sum_df.sum(), on="month_year")
real_estate_df
```
