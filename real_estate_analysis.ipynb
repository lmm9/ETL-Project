```python
# Data provided by Redfin, a national real estate brokerage
import pandas as pd
```

```python
# Store filepath in a variable
raw_data = "Resources/weekly_housing_market_rawdata_10.10.2020.csv"

# Read our Data file with the pandas library
raw_data_df = pd.read_csv(raw_data, encoding="utf-8")
raw_data_df.head()
```

```python
# Show just the header
len(raw_data_df)
```

```python
raw_data_df.columns
```

```python
columns1=[
    "region_name", 
    "region_type", 
    "period_begin", 
    "period_end",
    "total_homes_sold", 
    "total_new_listings", 
    "total_active_listings",
    "price_drops",
    "homes_delisted",
    "median_sale_price",
    "median_days_on_market",
    "months_of_supply",
    "age_of_inventory"]

weekly_df=raw_data_df.loc[raw_data_df["duration"] == "1 weeks",columns1]
weekly_df.head()
```

```python
len(weekly_df)
```

```python
weekly_df.dtypes
```

```python
#start a new DataFrame to cleanup
clean_df=weekly_df

#break out period_begin into month, day, year
clean_df[['month','day','year']] = clean_df.period_begin.str.split("/",expand=True)

#add leading 20 to year
clean_df['year'] ='20' + clean_df['year']

#replace month number with month name
clean_df['month']=clean_df['month'].replace(
{'1':'January',
'2':'February',
'3':'March',
'4':'April',
'5':'May',
'6':'June',
'7':'July',
'8':'August',
'9':'September',
'10':'October',
'11':'November',
'12':'December'})

#merge month and year for grouping
clean_df['month_year'] =clean_df['month'].astype(str) + " "+ clean_df['year']
clean_df.head()
```

```python
#break out region_name into city, and state
clean_df[['region','state']] = clean_df.region_name.str.split(", ",expand=True)

#only take first two characters of state
clean_df['state'] = clean_df['state'].str[:2]
clean_df.head()
```
